```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path1 = "ecommerce-behavior-data-from-multi-category-store/2019-Oct.csv"
file_path2 = "ecommerce-behavior-data-from-multi-category-store/2019-Nov.csv"

def read_large_file(file_path, chunksize=1000000):
    print(f"Membaca file: {file_path}")
    chunks = []
    for chunk in pd.read_csv(file_path, chunksize=chunksize):
        chunks.append(chunk)
        print(f"Chunk dengan {len(chunk)} baris selesai dibaca.")
    print(f"File {file_path} selesai dibaca.")
    return pd.concat(chunks, ignore_index=True)

print("Membaca file pertama...")
df1 = read_large_file(file_path1)

print("Membaca file kedua...")
df2 = read_large_file(file_path2)

print("\nMenggabungkan dataset...")
df = pd.concat([df1, df2], ignore_index=True)
print(f"Gabungan selesai. Total jumlah baris: {len(df)}")

print("\nPreview Dataset Gabungan:")
print(df.head())

print("\nInformasi Dataset:")
print(df.info())

print("\nStatistik Deskriptif:")
print(df.describe())

if 'event_type' in df.columns:
    print("\nVisualisasi distribusi event_type...")
    plt.figure(figsize=(8, 5))
    sns.countplot(x='event_type', data=df, order=df['event_type'].value_counts().index)
    plt.title("Distribusi Event Type")
    plt.xlabel("Tipe Event")
    plt.ylabel("Jumlah")
    plt.show()

if 'product_id' in df.columns:
    print("\nVisualisasi top 10 produk dengan event terbanyak...")
    top_products = df['product_id'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_products.index, y=top_products.values)
    plt.title("Top 10 Produk dengan Event Terbanyak")
    plt.xlabel("Product ID")
    plt.ylabel("Jumlah Event")
    plt.show()

output_path = "ecommerce_combined_data.csv"
df.to_csv(output_path, index=False)
print(f"\nDataset gabungan telah disimpan ke: {output_path}")
```
